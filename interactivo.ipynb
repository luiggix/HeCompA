```python
import matplotlib.pyplot as plt
import numpy as np
import ipywidgets as wd
```

```python
def graf(t, y):
    plt.plot(t, y)
    plt.show()
```

```python
def solver(w):
    t = np.linspace(0, 2* np.pi, 100)
    y = np.sin(w * t)
    graf(t,y)
```

```python
w = 4
solver(w)
```

```python
wd.interact(solver, w = wd.FloatSlider(min=0.0, max=10.0, step=1.0, value=-1.0, description='x0'))
```
