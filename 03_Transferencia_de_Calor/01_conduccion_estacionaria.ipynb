# Conducción de calor estacionaria 1D.

**Objetivos.**
Resolver un problema de conducción de calor estacionaria en 1D con diferentes parámetros y condiciones de frontera tipo Dirichlet.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def buildMatrix(N):
    # Matriz de ceros
    A = np.zeros((N,N))
	
    # Primer renglón
    A[0,0] = 2
    A[0,1] = -1
    # Renglones interiores
    for i in range(1,N-1):
        A[i,i] = 2
        A[i,i+1] = -1
        A[i,i-1] = -1
    # Último renglón
    A[N-1,N-2] = -1
    A[N-1,N-1] = 2
	
    return A
```

```python
# Parámetros físicos
L = 1.0
TA = 1
TB = 0
k = 1.0
S = 0.0
# Parámetros numéricos
N = 4
h = L / (N+1)
r = k / h**2

# Arreglo para almacenar la solución
T = np.zeros(N+2)
T[0]  = TA  # Frontera izquierda
T[-1] = TB  # Frontera derecha

# Lado derecho del sistema
b = np.zeros(N)
b[:] = S / r # Fuente o sumidero
b[0]  += T[0]   # Condición de frontera
b[-1] += T[-1]  # Condición de frontera

# Construcción de la matriz
A = buildMatrix(N)

# Solución del sistema lineal
T[1:N+1] = np.linalg.solve(A,b)
```

```python
# Impresión de la solución
print(T)
```

```python
# Graficación de la solución
x = np.linspace(0, L, N+2)
plt.figure(figsize=(10,7))
plt.plot(x, T, c='grey', lw=2.0)
plt.scatter(x, T, edgecolor='k', zorder= 10)
plt.xlabel('$x$')
plt.ylabel('$T$')
plt.grid()
plt.show()
```
