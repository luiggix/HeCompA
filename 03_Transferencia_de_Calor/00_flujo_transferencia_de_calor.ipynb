## Ejemplo: Horno industrial 
                                
La pared de un horno industrial está construida de ladrillos de arcilla refractaria
cuya conductividad térmica es de $1.7$ W/m$\cdot$K. En estado estacionario, durante su
operación, la temperatura interna tiene una temperatura de 1400 K, mientras que la
externa 1150 K. Si la pared es de $0.5$ m $\times$ $1.2$ m, ¿cuál es la razón de pérdida
de calor ($q_x$) a través de la pared?

<img src="./EjemploConduccion1D.png" width=600px/>


Tomado de: **Bergman, T.L. and Incropera, F.P. and DeWitt, D.P. and Lavine, A.S., Fundamentals of Heat and Mass Transfer, Wiley, 2011.**

</div>

## Paso 1. Definición de variables del problema

```python
T1 = 1400 # K
T2 = 1150 # K
k  = 1.7  # W/m K
L  = 0.15 # m
W = 1.2 # m
H = 0.5 # m
```

## Paso 2. Cálculo del flujo de calor.

$$
q_x^{\prime\prime} = \kappa \frac{T_1 - T_2}{L}
$$

```python
qx_pp = k * (T1 - T2) / L
print('Flujo de calor: {} [W/m^2]'.format(qx_pp))
```

## Paso 3. Transferencia de calor por conducción.

$$
q_x = q_x^{\prime\prime} \cdot A
$$

```python
qx = qx_pp * W * H
print('Transferencia de calor por conducción: {} [W]'.format(qx))
```
