$$
\begin{eqnarray}
\dfrac{dx}{dt} & = & \sigma (y - x) \\
\dfrac{dy}{dt} & = & x (\rho - z) - y\\
\dfrac{dz}{dt} & = & xy - \beta z \\
\end{eqnarray}
$$

* Inicia con una posición inicial $(0, 1, 1.05)$ y los siguientes valores de los parámetros $\sigma = 10$, $\beta = 8/3$, $\rho = 28$.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Definir parametros 
sigma = 10
beta =8/3
ro = 28
print(sigma,beta,ro)
```

```python
#definir condición inicial
x0 = 0
y0 = 1
z0 = 1.05
print (x0, y0, z0)
```

```python
Nt = 1000
ht = 0.01
print (Nt, ht)
```

```python
x= np.zeros(Nt+1)
y= np.zeros(Nt+1)
z= np.zeros(Nt+1)
print(x,y,z)
```

```python
x[0]=x0
y[0]=y0
z[0]=z0
print(x,y,z)
```

```python
for n in range(0,Nt):
    x[n+1] = x[n] + ht * sigma*(y[n]-x[n])
    y[n+1] = y[n] + ht * (x[n]*(ro -z[n])-y[n])
    z[n+1] = z[n] + ht * (x[n]*y[n]-beta*z[n])

print(x,y,z)
```

```python
# Graficación de las trayectorias en el espacio fase
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1], projection='3d')
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$z$')

# Graficamos la condicion inicial
ax.scatter(x[0], y[0], z[0], fc='green', ec='k', s = 50, zorder=10) 

# Graficamos la trayectoria
ax.plot(x, y, z, '-', lw='0.75')

# Graficamos la posición final
ax.scatter(x[-1], y[-1], z[-1], fc='red', ec='k', s = 50, zorder=5)

plt.show()
```
