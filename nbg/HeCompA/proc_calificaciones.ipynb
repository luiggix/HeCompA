```python
# Calculamos la calificación de cada alumno y de cada notebook y la ponemos en un archivo csv.

import pandas as pd
from nbgrader.api import Gradebook, MissingEntry

# Create the connection to the database
with Gradebook(database) as gb:

    grades = []

    # Loop over each assignment in the database
    for assignment in gb.assignments[2:3]:
        print(assignment.name)
        
        for n in assignment.notebooks:
            notebook_name = str(n).split(sep='/')[-1]
            notebook_name = assignment.name + '-' + notebook_name[:-2]
            print(notebook_name)
            
            # Loop over each student in the database
            for student in gb.students:
    
                # Create a dictionary that will store information about this student's
                # submitted assignment
                score = {}
                score['name'] = student.first_name
                score['lastname'] = student.last_name

                
                # Try to find the submission in the database. If it doesn't exist, the
                # `MissingEntry` exception will be raised, which means the student
                # didn't submit anything, so we assign them a score of zero.
                try:
                    submission = gb.find_submission(assignment.name, student.id)
                except MissingEntry:
                    score['score'] = 0.0
                else:
                    score['score'] = submission.score
                    
                score['max_score'] = assignment.max_score

                score['Dirección Email'] = student.id
                score['grades'] = round(10 * score['score'] / (assignment.max_score), 2)                

                grades.append(score)

    # Create a pandas dataframe with our grade information, and save it to disk
    grades = pd.DataFrame(grades).set_index(['name', 'lastname'])#.sort_index()
    ofile = 'grades-' + notebook_name + '.csv'
    grades.to_csv(ofile)

    # Print out what the grades look like
#    with open(ofile, 'r') as fh:
#        print(fh.read())
```

```python
# Imprimir la información de las calificaciones para posteriormente
# subirlas a Moodle con la opción de copiar y pegar.
print(grades.columns[2], '\t', grades.columns[3])
for e, g in zip(grades['Dirección Email'], grades['grades']):
    print(e, '\t', g)
```
